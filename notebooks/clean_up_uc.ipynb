# Clean Up UC OTel Tables

Truncates the OTel spans, logs, and metrics tables in Unity Catalog.

```python
CATALOG = "bx3"
SCHEMA = "otel_demo"

OTEL_SPANS_TABLE = f"{CATALOG}.{SCHEMA}.otel_spans_v2"
OTEL_LOGS_TABLE = f"{CATALOG}.{SCHEMA}.otel_logs_v2"
OTEL_METRICS_TABLE = f"{CATALOG}.{SCHEMA}.otel_metrics"

tables = [OTEL_SPANS_TABLE, OTEL_LOGS_TABLE, OTEL_METRICS_TABLE]
```

```python
print("Cleaning OTel UC target tables:\n")
for table in tables:
    before = spark.sql(f"SELECT COUNT(*) AS cnt FROM {table}").collect()[0]["cnt"]
    try:
        spark.sql(f"TRUNCATE TABLE {table}")
        method = "TRUNCATE TABLE"
    except Exception:
        spark.sql(f"DELETE FROM {table}")
        method = "DELETE FROM"
    after = spark.sql(f"SELECT COUNT(*) AS cnt FROM {table}").collect()[0]["cnt"]
    print(f"  {table}: {before} -> {after} ({method})")
```
